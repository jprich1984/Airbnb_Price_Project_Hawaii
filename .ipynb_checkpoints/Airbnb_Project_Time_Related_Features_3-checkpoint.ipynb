```python
def remove_outliers(data, column, method='iqr'):
    if method == 'iqr':
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    elif method == 'zscore':
        z_scores = np.abs(stats.zscore(data[column]))
        return data[z_scores < 3]
    else:
        raise ValueError("Method must be either 'iqr' or 'zscore'")
df_no_outliers = apf.remove_outliers(df, 'price', method='iqr')
apf.check_normality(df_no_outliers['log_price'])
```

# Create isHoliday

**Analyze different stretches of time within known holiday periods to determine what date ranges should be considered holiday periods.  Use t tests that compare the stretch's prices with prices of neighboring time periods**

```python
def analyze_holiday_period(df, start_date, end_date, min_days=1,p_threshold=0.05):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Calculate the length of the holiday period
    holiday_length = (end_date - start_date).days + 1

    # Define the outside window
    outside_window = df[
        ((df['date'] < start_date) & (df['date'] >= start_date - pd.Timedelta(days=holiday_length))) |
        ((df['date'] > end_date) & (df['date'] <= end_date + pd.Timedelta(days=holiday_length)))
    ]
    
    # Calculate mean and std for the outside window
    outside_mean = outside_window['log_price'].mean()
    outside_std = outside_window['log_price'].std()

    # Aggregate the data by date
    daily_avg = df.groupby('date')['log_price'].mean().reset_index()
    
    # Define the holiday period in the aggregated data
    holiday_period = daily_avg[(daily_avg['date'] >= start_date) & (daily_avg['date'] <= end_date)]
    
    best_period_tuple = None
    best_period_dict = None
    best_length = 0
    results = []

    # Check all possible sub-ranges within the holiday period
    for i in range(len(holiday_period)):
        for j in range(i + min_days - 1, len(holiday_period)):  # Start from i + min_days - 1 to include the min_days
            sub_range = holiday_period.iloc[i:j + 1]
            sub_prices = sub_range['log_price']
            sub_mean = sub_prices.mean()

            # Perform t-test on log transformed data
            t_stat, p_value = stats.ttest_1samp(sub_prices, outside_mean)
            is_significant = p_value < p_threshold and sub_mean > outside_mean

            sub_std = sub_prices.std()  # Calculate standard deviation for the sub-range
            log_prices_list = sub_prices.tolist()  # Convert the sub_prices Series to a list

            results.append({
                'start_date': sub_range['date'].min(),
                'end_date': sub_range['date'].max(),
                'mean': sub_mean,
                'std_dev': sub_std,  # Store standard deviation
                't_statistic': t_stat,
                'p_value': p_value,
                'is_significant': is_significant,
                'length': len(sub_range),
                'log_prices': log_prices_list  # Store the log prices for the range
            })
            # Update the best period if significant
            if is_significant and len(sub_range) > best_length:
                best_length = len(sub_range)
                best_period_tuple = (sub_range['date'].min(), sub_range['date'].max())
                best_period_dict = {
                    'start_date': sub_range['date'].min(),
                    'end_date': sub_range['date'].max(),
                    'mean': sub_mean,
                    'std_dev': sub_std,
                    't_statistic': t_stat,
                    'p_value': p_value,
                    'is_significant': is_significant,
                    'length': len(sub_range),
                    'log_prices': log_prices_list  # Store the log prices for the best period
                }

    return best_period_tuple, best_period_dict, results
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_aggregated_column_vs_date(df, y_column):
    """
    Groups the DataFrame by 'date', calculates the mean of the specified column,
    and plots the average values against the date.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    y_column (str): The name of the column to plot on the y-axis.

    Returns:
    None
    """
    # Ensure 'date' column is in datetime format
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'])

    # Group by 'date' and calculate the mean of the specified column
    aggregated_df = df.groupby('date')[y_column].mean().reset_index()

    # Plot the aggregated data
    plt.figure(figsize=(12, 6))
    plt.plot(aggregated_df['date'], aggregated_df[y_column], label=f'Average {y_column}')
    plt.title(f'Average {y_column} vs Date')
    plt.xlabel('Date')
    plt.ylabel(f'Average {y_column}')
    plt.legend()
    plt.grid(True)
    plt.show()


apf.plot_aggregated_column_vs_date(df_no_outliers, 'price')
```

# Aggregating the data by a time period

**The next section is about trying to figure out a good length of time for which to aggregate the data over.  If prices are not changing very
much from week to week we might be able to aggregate the data by month and make predictions for listings based on the month and the other features.  However, it may be necessary to aggregate by a smaller time window if hosts are chaning prices frequently.**

```python
df['week'] = df['date'].dt.to_period('W')
df_weekly = df.groupby(['listing_id', 'week'])['price'].agg(['mean', 'median', 'min', 'max']).reset_index()
def plot_time_aggregated_df(df,period,center='mean'):
    df_avg=df.groupby(period).agg({center:'mean'}).reset_index()
    fig, ax=plt.subplots()
    ax.scatter(x=df_avg[period].astype(int),y=df_avg[center],alpha=0.6)
    ax.set_xlabel(period)
    ax.set_ylabel('price')
    plt.suptitle(f'Price vs {period.capitalize()}')
    plt.tight_layout()
    plt.show()

plot_time_aggregated_df(df_weekly,'week')
```

```python
df['month'] = df['date'].dt.month
df_monthly = df.groupby(['listing_id', 'month'])['price'].agg(['mean', 'median', 'min', 'max']).reset_index()
plot_time_aggregated_df(df_monthly,'month')
```

**The average price tends to be different from week to week and from month to month.  We see that during some time periods, the price does not change very significantly on a weekly basis.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def count_price_changes(df, period):
    """
    Counts the number of times the price was changed from one period to the next for each listing.

    Parameters:
    df (pd.DataFrame): The original DataFrame containing 'listing_id', 'date', and 'price' columns.
    period (str): The period for aggregation ('week' or 'month').

    Returns:
    pd.DataFrame: A DataFrame with listing IDs and the count of price changes.
    """
    # Ensure date column is datetime
    df['date'] = pd.to_datetime(df['date'])
    
    # Calculate week and month
    df['week'] = df['date'].dt.isocalendar().week
    df['month'] = df['date'].dt.month
    
    # Aggregate by listing_id and period
    if period == 'week':
        df_agg = df.groupby(['listing_id', 'week'])['price'].mean().reset_index()
    elif period == 'month':
        df_agg = df.groupby(['listing_id', 'month'])['price'].mean().reset_index()
    else:
        raise ValueError("Period must be 'week' or 'month'")
    
    # Sort by listing_id and period
    df_agg = df_agg.sort_values(by=['listing_id', period])
    
    # Calculate price changes
    df_agg['price_change'] = df_agg.groupby('listing_id')['price'].diff().fillna(0)
    
    # Count non-zero price changes for each listing
    price_change_counts = df_agg.groupby('listing_id')['price_change'].apply(lambda x: (x != 0).sum()).reset_index()
    price_change_counts.columns = ['listing_id', f'{period}_price_changes']
    
    # Calculate total opportunities to change (total number of periods for each listing)
    total_opportunities = df_agg.groupby('listing_id')[period].nunique().reset_index()
    total_opportunities.columns = ['listing_id', f'{period}_opportunities']
    
    # Merge the counts with the opportunities
    price_change_counts = price_change_counts.merge(total_opportunities, on='listing_id')
    
    return price_change_counts

# Assuming 'df' is your original DataFrame
# Count price changes for weekly and monthly periods
weekly_price_changes = count_price_changes(df, 'week')
monthly_price_changes = count_price_changes(df, 'month')

# Merge the results
price_change_summary = pd.merge(weekly_price_changes, monthly_price_changes, on='listing_id', how='outer', suffixes=('_weekly', '_monthly'))

# Calculate the total number of changes and opportunities
total_weekly_changes = price_change_summary['week_price_changes'].sum()
total_weekly_opportunities = price_change_summary['week_opportunities'].sum()
total_monthly_changes = price_change_summary['month_price_changes'].sum()
total_monthly_opportunities = price_change_summary['month_opportunities'].sum()

# Calculate the percentages
weekly_change_percentage = (total_weekly_changes / total_weekly_opportunities) * 100
monthly_change_percentage = (total_monthly_changes / total_monthly_opportunities) * 100

print(f"Total weekly changes: {total_weekly_changes}")
print(f"Total weekly opportunities: {total_weekly_opportunities}")
print(f"Weekly change percentage: {weekly_change_percentage:.2f}%")

print(f"Total monthly changes: {total_monthly_changes}")
print(f"Total monthly opportunities: {total_monthly_opportunities}")
print(f"Monthly change percentage: {monthly_change_percentage:.2f}%")

# Visualize the distribution of price changes
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(price_change_summary['week_price_changes'], bins=20)
plt.title('Distribution of Weekly Price Changes')
plt.xlabel('Number of Changes')
plt.ylabel('Number of Listings')

plt.subplot(1, 2, 2)
plt.hist(price_change_summary['month_price_changes'], bins=20)
plt.title('Distribution of Monthly Price Changes')
plt.xlabel('Number of Changes')
plt.ylabel('Number of Listings')

plt.tight_layout()
plt.show()
```

**Enough listings are changing their prices by week to justify aggregating by this time period.**

```python
import pandas as pd

import re

def look_at_listing_pool_amenities(row, pattern):
    if isinstance(row['amenities'], list):
        for amenity in row['amenities']:
            if isinstance(amenity, str):
                if re.search(pattern, amenity, re.IGNORECASE):
                    return True
    return False


def get_pool_info(data):
    df=data.copy()
    new_cols=['hasPool','hasPrivatePool','hasHotTub','hasPrivateHotTub']
    df[new_cols[0]] = df.apply(lambda x: look_at_listing_pool_amenities(x, r'pool'), axis=1)
    df[new_cols[1]] = df.apply(lambda x: look_at_listing_pool_amenities(x, r'private pool'), axis=1)
    df[new_cols[2]] = df.apply(lambda x: look_at_listing_pool_amenities(x, r'hot tub'), axis=1)
    df[new_cols[3]] = df.apply(lambda x: look_at_listing_pool_amenities(x, r'private hot tub'), axis=1)
    fig, ax=plt.subplots(len(new_cols),figsize=(20,14))
    for i,col in enumerate(new_cols):
        grouped=df.groupby(col).agg({'price':'mean'}).reset_index()
        sns.barplot(x=col,y='price',data=grouped,ax=ax[i])
        ax[i].set_xlabel(f'{col}')
        ax[i].set_ylabel(f'Average Price')
    fig.suptitle('Average Price by Amenity type')
    plt.tight_layout()
    plt.show()
    return df
pool_info=get_pool_info(df_aggregated)
```

```python
APPLY=False
import swifter
def hasPrivatePoolorTub(amenities):
    pattern1 = r"private pool"
    pattern2 = r"private hot tub"
    pool=False
    hot_tub=False
    for text in amenities:
        if isinstance(text, str):
            if re.search(pattern1, text, re.IGNORECASE):
                pool=True
            elif re.search(pattern2, text, re.IGNORECASE):
                hot_tub=True
            if pool and hot_tub:
                break
    return pd.Series({'hasPool': pool, 'hasHotTub': hot_tub})
if APPLY:
    df_aggregated[['hasPool','hasHotTub']]=df_aggregated['amenities'].apply(hasPrivatePoolorTub)
```

```python
#Vectorized Approach
# Convert lists to strings
df_aggregated['amenities_str'] = df_aggregated['amenities'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# Vectorized approach using str.contains
df_aggregated['hasPool'] = df_aggregated['amenities_str'].str.contains(r"private pool", case=False, na=False)
df_aggregated['hasHotTub'] = df_aggregated['amenities_str'].str.contains(r"private hot tub", case=False, na=False)

print(df_aggregated[['amenities', 'hasPool', 'hasHotTub']])
```
