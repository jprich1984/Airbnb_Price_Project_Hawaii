```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Project Objective: 
## &emsp;Create a model that can predict the price of an airbnb unit based on the time of year and listing characteristics.

# Notebook Objective: 
## &emsp;Remove data points that contain potentially mislabeled variables, and/or remove data points that are not consistent with the expected relationships that we wish our model to capture.

**Throughout this process I discovered many listings that had mislabeled prices (e.g. The price that was in the data set was inconsistent with the price that was listed on the airbnb website).  Furthermore, I found instances of listings with extremely unreasonable prices.  Interestingly, these listings often did not have many reviews.  Lastly, some of the listings are very expensive for reasons that cannot be explained by the current features.  This notebook seeks to thoughtfully choose the data that will be used to train the model.**

```python
df=pd.read_csv("airbnb_project_clean5.csv")
columns=[col for col in df.columns if not col.startswith('Unnamed')]
df=df[columns]
df.head()
df['log_price']=np.log(df['price'])
df.info()
```

# Use a linear model to determine the listings that should stay in the data.  Specifically, filter out listings where the  price is 50% higher/lower than the predicted price. 

```python
df=pd.get_dummies(df,columns=['property_group','resortArea','neighbourhood_cleansed'],drop_first=True)
features=[col for col in df.columns if col not in ['listing_id','log_price','listing_url','latitude','longitude','geometry','property_type','price_diff','number_of_reviews','amenities_length']]
num_features=['price','bathrooms','bedrooms','accommodates','month','distanceToOcean','isHoliday']
cat_features=[col for col in df.columns if 'resort' in col]+[col for col in df.columns if 'neighbourhood' in col]+[col for col in df.columns if 'property_group' in col]
features=features+cat_features
```

**Features that will be used for linear model**

```python
df[features].info()
```

```python
def get_outliers(data,col,factor):
    quarts=data[col].quantile([0.25,0.75])
    iqr=quarts.loc[0.75]-quarts.loc[0.25]
    upper_boundary=quarts.loc[0.75]+iqr*factor
    lower_boundary=quarts.loc[0.25]-iqr*factor
    return data[(data[col]<lower_boundary)|(data[col]>upper_boundary)]
price_outliers=get_outliers(df,'log_price',1.5)
price_outliers.info()
price_outliers.sort_values(by='price', ascending=False)[['price','bedrooms']]
```

# The iqr method Identifies roughly 11,000 data points that are outliers with respect to price.

*Ensure numeric data type*

```python
from pandas.api.types import is_numeric_dtype
features=[col for col in features if col!='price']
features=[col for col in features if is_numeric_dtype(df[col])]
```

*Create dummy columns for month*

```python
features=[col for col in features if col!='month']
df=pd.get_dummies(df,columns=['month'])
```

*Convert bool to integer. The stats model that is used requires integer datatypes*

```python

bool_columns = df.select_dtypes(include=['bool']).columns.tolist()
# Convert boolean columns to integers
for col in bool_columns:
    df[col] = df[col].astype(int)
# Select features that are monotonically related to the target
num_features = ['bedrooms', 'bathrooms', 'accommodates', 'distanceToOcean','hasPool','hasHotTub']
features=num_features+[col for col in df.columns if 'month' in col]+[col for col in df.columns if 'resort' in col]
df[features].info()
```

**Fit linear model**

```python
import pandas as pd
import statsmodels.api as sm
X = df[features]
y = df['log_price']  
df.dropna(subset=features + ['log_price'], inplace=True)
X = df[features]
y = df['log_price']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
df['predicted_price'] = model.predict(X)
df['residuals'] = df['log_price'] - df['predicted_price']
# Identify outliers based on residuals
threshold = 1.5 * df['residuals'].std()  # You can adjust this threshold
outliers = df[(df['residuals'] > threshold) | (df['residuals'] < -threshold)]
# Display identified outliers
print("Identified Outliers:")
print(outliers[['log_price', 'predicted_price', 'residuals']])
print(model.summary())
```

# Look at some rows where the price was significantly different from the predicted price.

```python
resort_cols=[col for col in df.columns if 'resort' in col]
outliers[['price','bedrooms','latitude','longitude','week','accommodates','bathrooms','listing_url','number_of_reviews','distanceToOcean']+resort_cols].sample(20)
```

**The number of reviews stands out. Specifically, it seems like a lot of these listings have a small number of reviews**

```python
fig, ax=plt.subplots()
ax.scatter(x='number_of_reviews',y='price',data=df)
ax.set_xlabel('number_of_reviews')
ax.set_ylabel('Price')
fig.suptitle('number_of_reviews Vs Price')

plt.show()
```

# Remove Rows

```python

df_cleaned = df[~df.index.isin(outliers.index)]

# Optionally, reset the index of the cleaned DataFrame
df_cleaned.reset_index(drop=True, inplace=True)

# Display the cleaned DataFrame
print("Cleaned DataFrame without outliers:")
print(df_cleaned)
```

**Remove additional high priced units and units with a low number of reviews**

```python
price_outliers=get_outliers(df_cleaned,'log_price',1.5)
price_outliers.info()
price_outliers.sort_values(by='price', ascending=False)[['price','bedrooms','listing_url','accommodates','number_of_reviews']].to_csv('outliers.csv')
drops=[]
for ind, row in price_outliers.iterrows():
    if row['number_of_reviews']<4:
        drops.append(ind)
    if row['price']<60:
        drops.append(ind)
    if row['price']>2500:
        drops.append(ind)
df_cleaned=df_cleaned[~df_cleaned.index.isin(drops)]

df_cleaned.reset_index(drop=True, inplace=True)
```

```python
print('Percentage of Data Set Removed: ', 100*(len(df)-len(df_cleaned))/len(df))
```

```python
fig, ax=plt.subplots()
ax.scatter(x='number_of_reviews',y='price',data=df_cleaned)
ax.set_xlabel('number_of_reviews')
ax.set_ylabel('price')
fig.suptitle('number_of_reviews Vs Price')
plt.show()
```

# Write Cleaned Dataframe to csv

```python
df_cleaned.drop(columns=['predicted_price','residuals'],inplace=True)
df_cleaned.to_csv("airbnbHawaii_clean6.csv",index=False)
df_cleaned.info()
```
