# Train-Test split
**Split the data by listing.  This ensures that the test set will contain truly unseen data.  If you were to split on time or using sklearn.train_test_split the same listings would be in the training and testing sets, just for different dates.**

```python
import random
def train_test_split_airbnb(df,train_size):
    random.seed(2)
    listings=df['listing_id'].unique()
    indices = list(range(len(listings)))
    train_indices=random.sample(indices, int(train_size*len(indices)))
    test_indices=[ ind for ind in indices if ind not in train_indices]
    train_listings=[listings[ind] for ind in train_indices]
    test_listings=[listings[ind] for ind in test_indices]
    train=df[df['listing_id'].isin(train_listings)]
    test=df[df['listing_id'].isin(test_listings)]
    return train,test
                                 
    
train, test=train_test_split_airbnb(df,0.8)

X_train=train[features]
y_train=train['price']
X_test=test[features]
y_test=test['price']
```

```python
def RMSE(y_true,y_pred):
    return np.sqrt(sum((y_true-y_pred)**2)/len(y_true))
def get_predictions(model,X):
    return model.predict(X)

def get_RMSE(model,X,y):
    preds=get_predictions(model,X)
    return RMSE(y,preds)
def symmetric_mean_absolute_percentage_error(y_true, y_pred):
    """
    Calculate Symmetric Mean Absolute Percentage Error (SMAPE).
    This metric is more robust to outliers and avoids division by zero.
    
    Args:
    y_true (array-like): Array of true values
    y_pred (array-like): Array of predicted values
    
    Returns:
    float: SMAPE value
    """
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(2.0 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred))) * 100
def get_percentage_based_metric(model, X,y):
    preds=get_predictions(model, X)
    return symmetric_mean_absolute_percentage_error(y,preds)
```

# Custom Cross Validation Function

```python


def custom_cross_validation(df, n_splits=5,seed=42):
    np.random.seed(seed)
    listings = df['listing_id'].unique()
    n_listings = len(listings)
    
    random.shuffle(listings)
    
    fold_size = n_listings // n_splits
    
    folds = [list(listings[i * fold_size:(i + 1) * fold_size]) for i in range(n_splits)]
    leftover = listings[n_splits * fold_size:]
    for i, listing in enumerate(leftover):
        folds[i].append(listing)

    for i in range(n_splits):
        val_listings = folds[i] 
        train_listings = [listing for j in range(n_splits) if j != i for listing in folds[j]]
        train = df[df['listing_id'].isin(train_listings)]
        val = df[df['listing_id'].isin(val_listings)]
        
        yield train, val
```

```python


def get_combinations_xg(param_grid):
    # Generate all combinations of parameters
    param_combinations = list(itertools.product(
        param_grid['n_estimators'],
        param_grid['max_depth'],
        param_grid['learning_rate'],
        param_grid['subsample'],
        param_grid['colsample_bytree'],
        param_grid['gamma']
    ))
    return param_combinations
def get_combinations_exTrees(param_grid):
    # Generate all combinations of parameters
    param_combinations = list(itertools.product(
    param_grid['n_estimators'],
    param_grid['max_depth'],
    param_grid['min_samples_split'],
    param_grid['min_samples_leaf'],
    param_grid['max_features']
))
    return param_combinations
def get_combinations_rf(param_grid):
    # Generate all combinations of parameters
    param_combinations = list(itertools.product(
        param_grid['n_estimators'],
        param_grid['max_depth'],
        param_grid['min_samples_split'],
        param_grid['min_samples_leaf'],
        param_grid['max_features']
    ))
    return param_combinations


def get_combinations_rf(param_grid):
    # Generate all combinations of parameters
    param_combinations = list(itertools.product(
    param_grid['n_estimators'],
    param_grid['max_depth'],
    param_grid['min_samples_split'],
    param_grid['min_samples_leaf'],
    param_grid['max_features']
))
    return param_combinations

def apply_grid_search(param_grid, param_combinations,filename, model_type, mode='a'):
    # Open CSV for writing results
    with open(filename, mode, newline='') as f:
        writer = csv.writer(f)
        if mode=='w':
            writer.writerow([*param_grid.keys(), 'RMSE', 'R2'])  # Added RMSE and R2 to header
    
        # Outer loop: iterate over parameter combinations
        for grid in param_combinations:
            RMSE = []
            R2 = []
    
            # Inner loop: perform custom cross-validation
            for train, val in custom_cross_validation(df, n_splits=5):
                X_train = train[features]
                y_train = train['price']
                X_val = val[features]
                y_val = val['price']
    
                # Initialize the XGBoost model with the current parameter combination
                if model_type=='XG':
                    model = xgb.XGBRegressor(
                        n_estimators=grid[0],          # Number of trees in the forest
                        max_depth=grid[1],              # Maximum depth of the tree
                        learning_rate=grid[2],          # Step size shrinkage used in update to prevent overfitting
                        subsample=grid[3],              # Proportion of samples to use for training each tree
                        colsample_bytree=grid[4],       # Proportion of features to use for training each tree
                        gamma=grid[5]                   # Minimum loss reduction required to make a further partition
                    )
                    WRITE_MODE=0
                elif model_type =='ExTrees':
                    model = ExtraTreesRegressor(
                            n_estimators=grid[0],          
                            max_depth=grid[1],              
                            min_samples_split=grid[2],       
                            min_samples_leaf=grid[3],       
                            max_features=grid[4],                          
                    random_state=42)
                    WRITE_MODE=1
                elif model_type=='RF':
                    model = RandomForestRegressor(
                                n_estimators=grid[0],          # Number of trees in the forest
                                max_depth=grid[1],              # Maximum depth of the tree
                                min_samples_split=grid[2],      # Minimum samples required to split an internal node
                                min_samples_leaf=grid[3],       # Minimum samples required to be at a leaf node
                                max_features=grid[4]            # Number of features to consider when looking for the best split
                )
                    WRITE_MODE=2
                else:
                    print('Model Type Not Found')
                    
                    
    
                # Fit the model and make predictions
                model.fit(X_train, y_train)
                y_pred = model.predict(X_val)
    
                # Calculate RMSE and R^2
                rmse = np.sqrt(mean_squared_error(y_val, y_pred))
                r2 = r2_score(y_val, y_pred)
    
                # Append results
                RMSE.append(rmse)
                R2.append(r2)
    
            # Print average metrics for the current parameter combination
            print(f'Average RMSE: {np.mean(RMSE):.4f}, Average R²: {np.mean(R2):.4f}')
            if WRITE_MODE==0:
            # Write the average RMSE and R^2 for the current parameter combination
                writer.writerow([grid[0], grid[1], grid[2], grid[3], grid[4], grid[5], np.mean(RMSE), np.mean(R2)])
            elif WRITE_MODE==1:
                writer.writerow([grid[0],grid[1],grid[2],grid[3],grid[4],rmse,r2])
            elif WRITE_MODE==2:
                writer.writerow([grid[0], grid[1], grid[2], grid[3], grid[4], np.mean(RMSE), np.mean(R2)])

```

# Use Early Stopping

```python
def early_stopping_cv(df, model_params, early_stopping_rounds_list, n_splits=5):
    non_features = ['price', 'inResort', 'number_of_reviews', 'geometry', 'week', 'listing_id', 'log_price', 'listing_url', 'missing_rating', 'price_per_accommodation', 'price_per_bedroom', 'price_per_review', 'property_type', 'geometry', 'anomaly_score', 'price_per_accommodation', 'price_per_bedroom', 'price_per_review', 'amenities_length']
    features = [col for col in df.columns if col not in non_features]
    
    results = []
    seed = 42

    for early_stopping_rounds in early_stopping_rounds_list:
        fold_scores = []
        for train, val in custom_cross_validation(df, n_splits=n_splits):
            X_train, y_train = train[features], train['price']
            X_val, y_val = val[features], val['price']

            model = xgb.XGBRegressor(**model_params, early_stopping_rounds=early_stopping_rounds)
            model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)

            y_pred = model.predict(X_val)
            mse = mean_squared_error(y_val, y_pred)
            r2 = r2_score(y_val, y_pred)
            fold_scores.append((mse, r2))

        avg_mse = np.mean([score[0] for score in fold_scores])
        avg_r2 = np.mean([score[1] for score in fold_scores])
        results.append((early_stopping_rounds, avg_mse, avg_r2))

    return results, model.best_iteration

# Set your model parameters
model_params = {
    'n_estimators': 1000,  # Set this high, early stopping will prevent all from being built
    'subsample': 0.7,
    'colsample_bytree': 0.6,
    'learning_rate': 0.1,
    'max_depth': 8,
    'gamma': 0
    # Add other parameters as needed
}

# List of early stopping rounds to test
early_stopping_rounds_list = [10, 20, 30, 50, 70, 100]

# Perform cross-validation with early stopping
results, num_trees = early_stopping_cv(df, model_params, early_stopping_rounds_list)
print(f'Number of trees {num_trees}')
# Plot results
plt.figure(figsize=(12, 6))
early_stopping_rounds, mses, r2s = zip(*results)

plt.subplot(1, 2, 1)
plt.plot(early_stopping_rounds, mses, marker='o')
plt.title('MSE vs Early Stopping Rounds')
plt.xlabel('Early Stopping Rounds')
plt.ylabel('Mean Squared Error')

plt.subplot(1, 2, 2)
plt.plot(early_stopping_rounds, r2s, marker='o')
plt.title('R² vs Early Stopping Rounds')
plt.xlabel('Early Stopping Rounds')
plt.ylabel('R² Score')

plt.tight_layout()
plt.savefig('early_stopping_results.png')
plt.show()

# Print results
for early_stopping_rounds, mse, r2 in results:
    print(f"Early Stopping Rounds: {early_stopping_rounds}")
    print(f"Average MSE: {mse:.4f}")
    print(f"Average R²: {r2:.4f}")
    print()

# Find best early stopping rounds
best_early_stopping_rounds = min(results, key=lambda x: x[1])[0]
print(f"Best Early Stopping Rounds: {best_early_stopping_rounds}")
```

# Cross Validated Permutation Feature Importance

**Use permutation feature importance on multipe versions of train/test.  The results of these models is definitely different depending on the version of the training/test set that is used.  This implementation uses 5 fold cross validation and 12 permutations.**

```python
from xgboost import XGBRegressor

def permutation_importance_cv(model, X, y, feature, n_repeats=12):
    np.random.seed(2)
    baseline_mse = mean_squared_error(y, model.predict(X))
    importances = []
    for _ in range(n_repeats):
        X_permuted = X.copy()
        X_permuted[feature] = np.random.permutation(X_permuted[feature])
        permuted_mse = mean_squared_error(y, model.predict(X_permuted))
        importances.append(permuted_mse - baseline_mse)
    return np.mean(importances)

def cross_validated_feature_importance(df, model_class, model_params, n_splits=5):
    feature_importances = []
    non_features=['price','inResort','number_of_reviews','geometry','week','listing_id','log_price','listing_url','missing_rating','price_per_accommodation','price_per_bedroom','price_per_review','property_type','geometry','anomaly_score','price_per_accommodation','price_per_bedroom','price_per_review','amenities_length']
    
    features=[col for col in df.columns if col not in non_features]
    seed=2
    for train, val in custom_cross_validation(df, seed,n_splits=n_splits):
        X_train = train[features]
        y_train = train['price']
        X_val = val[features]
        y_val = val['price']
        
        # Train the model
        model = model_class(**model_params)
        model.fit(X_train, y_train)
        
        # Calculate feature importance for this fold
        fold_importances = []
        for feature in X_val.columns:
            importance = permutation_importance_cv(model, X_val, y_val, feature)
            fold_importances.append({'feature': feature, 'importance': importance})
        
        feature_importances.append(pd.DataFrame(fold_importances))
    
    # Aggregate importances across folds
    all_importances = pd.concat(feature_importances)
    mean_importances = all_importances.groupby('feature')['importance'].mean().sort_values(ascending=False)
    std_importances = all_importances.groupby('feature')['importance'].std()
    
    return mean_importances, std_importances

# Set your model parameters
best_model_params = {
    'n_estimators': 350,
    'subsample':0.7,
    'colsample_bytree':0.6,
    'learning_rate': 0.05,
    'max_depth': 8,
    'gamma':0
    # Add other parameters as needed
}

# Perform cross-validated feature importance
mean_importances, std_importances = apf.cross_validated_feature_importance(train, XGBRegressor, best_model_params)

# Plot feature importances
plt.figure(figsize=(12, 10))
mean_importances.plot(kind='bar', yerr=std_importances, capsize=4)
plt.title('Cross-Validated Permutation Feature Importance')
plt.xlabel('Features')
plt.ylabel('Mean Increase in MSE')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('FeatureImportance_10fold.png')
plt.show()

# Print feature importances
print("Cross-Validated Feature Importances:")
for feature, importance in mean_importances.items():
    print(f"{feature}: {importance:.4f} ± {std_importances[feature]:.4f}")
```

# Look at how the predicted and actual price adjust over time. 
**&emsp;The time column in the plots is the weeks since september 9th.  So time 0 is september 9th and time 20 would be 20 weeks after september 9th.**

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_price_comparison(df, listing_id,detail="",plot_single=False):
    # Filter the dataframe for the specified listing_id
    single_listing = df[df['listing_id'] == listing_id]
    
    # Sort the dataframe by week to ensure proper ordering in the plot
    single_listing = single_listing.sort_values('time')

    if plot_single:
    # Create a figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))
        
        # Plot week vs price
        ax1.plot(single_listing['time'], single_listing['price'], marker='o')
        ax1.set_title(f'Week vs Price for Listing ID: {listing_id}')
        ax1.set_xlabel('Week')
        ax1.set_ylabel('Price')
        
        # Plot week vs predicted_price
        ax2.plot(single_listing['time'], single_listing['predicted_price'], marker='o', color='red')
        ax2.set_title(f'Week vs Predicted Price for Listing ID: {listing_id}')
        ax2.set_xlabel('time')
        ax2.set_ylabel('Predicted Price')
        
        # Adjust layout for the separate plots
        plt.tight_layout()
        plt.show()
    
    # Create a single plot comparing actual and predicted prices
    plt.figure(figsize=(10, 8))
    plt.plot(single_listing['time'], single_listing['price'], marker='o', label='Actual Price')
    plt.plot(single_listing['time'], single_listing['predicted_price'], marker='o', color='red', label='Predicted Price')
    plt.title(f"Time vs Price (Actual and Predicted) Bedrooms: {single_listing['bedrooms'].iloc[0]} Ocean Distance: {single_listing['distanceToOcean'].iloc[0]}")

    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.tight_layout()
    plt.show()
listing_ids=sample['listing_id']
# Example usage:
plot_price_comparison(test, listing_ids.iloc[0])
```

**The Actual price and predicted price have similar behavior over time.  Interestingly during the fall the predicted price is relatively close to the actual price.  As time goes by into the winter and spring the listing raises their price quite a bit and the model does not adjust it's predicted price by as much.**

```python
plot_price_comparison(test, listing_ids.iloc[6])
```

**We see that this listing does not adjust their price quite as much.  The predicted price generally follows the direction of the price changes that actually do occur**

# Extentions: Quantile Regressor

```python

optimized_params = {
    'n_estimators': 350,
    'max_depth': 8,
    'learning_rate': 0.05,
    'subsample': 0.70,
    'colsample_bytree': 0.6,
    'gamma': 0.00
}

# Function to create a model with optimized parameters for a specific quantile
def create_quantile_model(quantile):
    params = optimized_params.copy()
    params['objective'] = 'reg:quantileerror'
    params['quantile_alpha'] = quantile
    return XGBRegressor(**params)

# Create and train models for different quantiles
model_lower = create_quantile_model(0.025)
model_median = create_quantile_model(0.5)
model_upper = create_quantile_model(0.975)

model_lower.fit(X_train, y_train)
model_median.fit(X_train, y_train)
model_upper.fit(X_train, y_train)

# Make predictions
y_pred_lower = model_lower.predict(X_test)
y_pred_median = model_median.predict(X_test)
y_pred_upper = model_upper.predict(X_test)

# Create a DataFrame with predictions and prediction intervals
results_df = pd.DataFrame({
    'Actual': y_test,
    'Prediction': y_pred_median,
    'Lower_PI': y_pred_lower,
    'Upper_PI': y_pred_upper
})

# Display the first few rows
print(results_df.head(10))

# Calculate coverage
coverage = np.mean((y_test >= y_pred_lower) & (y_test <= y_pred_upper))
print(f"\nCoverage: {coverage:.2%}")

# Calculate average PI width as a percentage of the predicted value
avg_pi_width_percentage = np.mean((y_pred_upper - y_pred_lower) / y_pred_median) * 100
print(f"Average PI width: {avg_pi_width_percentage:.2f}% of predicted value")

# Calculate MAPE
mape = np.mean(np.abs((y_test - y_pred_median) / y_test))
print(f"Mean Absolute Percentage Error: {mape:.2%}")
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'model' is your original XGBoost model
original_predictions = model.predict(X_test)

# Calculate RMSE for original model
rmse_original = np.sqrt(mean_squared_error(y_test, original_predictions))

# Calculate RMSE for median predictions from quantile regression
rmse_median = np.sqrt(mean_squared_error(y_test, y_pred_median))

# Calculate R2 score for original model
r2_original = r2_score(y_test, original_predictions)

# Calculate R2 score for median predictions from quantile regression
r2_median = r2_score(y_test, y_pred_median)

print(f"RMSE for original XGBoost model: {rmse_original:.2f}")
print(f"RMSE for median prediction (quantile regression): {rmse_median:.2f}")
print(f"R2 score for original XGBoost model: {r2_original:.4f}")
print(f"R2 score for median prediction (quantile regression): {r2_median:.4f}")

# Calculate the difference in RMSE
rmse_difference = abs(rmse_original - rmse_median)
print(f"Absolute difference in RMSE: {rmse_difference:.2f}")

# Calculate percentage difference in RMSE
rmse_percent_diff = (rmse_difference / rmse_original) * 100
print(f"Percentage difference in RMSE: {rmse_percent_diff:.2f}%")

# Calculate the difference in R2
r2_difference = abs(r2_original - r2_median)
print(f"Absolute difference in R2: {r2_difference:.4f}")

# Calculate percentage difference in R2
r2_percent_diff = (r2_difference / r2_original) * 100
print(f"Percentage difference in R2: {r2_percent_diff:.2f}%")

# Additional comparison
comparison_df = pd.DataFrame({
    'Actual': y_test,
    'Original_Prediction': original_predictions,
    'Median_Prediction': y_pred_median
})

print("\nFirst few rows of predictions:")
print(comparison_df.head(10))

# Calculate correlation between predictions
correlation = np.corrcoef(original_predictions, y_pred_median)[0, 1]
print(f"\nCorrelation between predictions: {correlation:.4f}")
```

```python
# Assuming 'test' is your original test DataFrame
# and X_test contains the same rows in the same order as 'test'

# Add predictions and intervals to the test DataFrame
test['Xg_Prediction'] = original_predictions
test['Median_Xg_Prediction'] = y_pred_median
test['Lower_PI'] = y_pred_lower
test['Upper_PI'] = y_pred_upper

# Calculate the width of the prediction interval
test['PI_Width'] = test['Upper_PI'] - test['Lower_PI']

# Calculate the percentage width of the prediction interval relative to the median prediction
test['PI_Width_Percentage'] = (test['PI_Width'] / test['Median_Xg_Prediction']) * 100

# Calculate the error (difference between actual and predicted)
test['Error'] = test['price'] - test['Median_Xg_Prediction']

# Calculate the absolute percentage error
test['Absolute_Percentage_Error'] = (abs(test['Error']) / test['price']) * 100

# Check if the actual price falls within the prediction interval
test['Within_PI'] = (test['price'] >= test['Lower_PI']) & (test['price'] <= test['Upper_PI'])

# Display the first few rows of the updated DataFrame
print(test.head())

# Display summary statistics for the new columns
summary_stats = test[['Xg_Prediction', 'Median_Xg_Prediction', 'Lower_PI', 'Upper_PI', 
                      'PI_Width', 'PI_Width_Percentage', 'Error', 'Absolute_Percentage_Error']].describe()
print("\nSummary Statistics:")
print(summary_stats)

# Calculate the percentage of actual prices falling within the prediction interval
coverage_percentage = test['Within_PI'].mean() * 100
print(f"\nPercentage of actual prices within the prediction interval: {coverage_percentage:.2f}%")

# Optional: Save the updated DataFrame to a CSV file
# test.to_csv('test_with_predictions.csv', index=False)
```
