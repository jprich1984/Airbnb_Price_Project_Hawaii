# Compare the average price of the various property_types (non-parametric t-test).
*Look for property_types that are similar/different in terms of price.  Use Kruska-Wallis H-test to compare mean prices.*

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.stats.multicomp as mc

def compare_property_types(df):
    # Group the data by property type
    grouped = df.groupby('property_type')
    
    # Perform Kruskal-Wallis H-test
    property_types = []
    prices = []
    for name, group in grouped:
        property_types.append(name)
        prices.append(group['price'])
    
    h_statistic, p_value = stats.kruskal(*prices)
    
    print(f"Kruskal-Wallis H-test result:")
    print(f"H-statistic: {h_statistic}")
    print(f"p-value: {p_value}")
    
    if p_value < 0.05:
        print("\nPerforming post-hoc pairwise comparisons:")
        all_prices = [item for sublist in prices for item in sublist]
        all_groups = [i for i, group in enumerate(prices) for _ in group]
        
        tukey = mc.pairwise_tukeyhsd(all_prices, all_groups)
        
        tukey_df = pd.DataFrame(data=tukey._results_table.data[1:], columns=tukey._results_table.data[0])
        
        # Add property type names to the DataFrame
        tukey_df['group1_name'] = tukey_df['group1'].apply(lambda x: property_types[int(x)])
        tukey_df['group2_name'] = tukey_df['group2'].apply(lambda x: property_types[int(x)])
        
        # Convert 'reject' column to boolean
        tukey_df['reject'] = tukey_df['reject'].astype(bool)
        
        return tukey_df
    else:
        print("No significant differences found among property types.")
        return None

# Assuming your DataFrame is named 'df'
tukey_results = compare_property_types(df)

if tukey_results is not None:
    print(tukey_results)
```

```python
def print_tukey_pair(name1,name2):
    results=tukey_results[(tukey_results['group1_name']==name1)&(tukey_results['group2_name']==name2)]
    return results
print_tukey_pair('Entire home','Treehouse')
```

# Create smaller property groups

```python
import pandas as pd
from itertools import combinations

def create_significant_diff_dict(tukey_df):
    # Create a dictionary to store significant differences
    property_keys = list(set(tukey_df['group1_name'].tolist() + tukey_df['group2_name'].tolist()))
    significant_diff = {key: [] for key in property_keys}
    
    for _, row in tukey_df.iterrows():
        if row['reject']:
            significant_diff[row['group1_name']].append(row['group2_name'])
            significant_diff[row['group2_name']].append(row['group1_name'])
    
    return significant_diff

def group_property_types(significant_diff):
    # Sort property types by the number of significant differences (descending order)
    sorted_property_types = sorted(significant_diff.keys(), key=lambda k: len(significant_diff[k]), reverse=True)
    
    # Initialize groups
    groups = []
    
    while sorted_property_types:
        # Start a new group with the property type that has the most significant differences
        current_group = [sorted_property_types.pop(0)]
        
        # Check remaining property types to see if they can be added to the current group
        i = 0
        while i < len(sorted_property_types):
            prop = sorted_property_types[i]
            # Check if prop can be added to the current group
            if all(prop not in significant_diff[member] for member in current_group):
                current_group.append(prop)
                sorted_property_types.pop(i)
            else:
                i += 1
        
        groups.append(current_group)
    
    return groups

# Function to validate groups
def validate_groups(groups, tukey_df):
    for group in groups:
        for pt1, pt2 in combinations(group, 2):
            result = tukey_df[(tukey_df['group1_name'] == pt1) & (tukey_df['group2_name'] == pt2) |
                              (tukey_df['group1_name'] == pt2) & (tukey_df['group2_name'] == pt1)]
            if not result.empty and result['reject'].iloc[0]:
                print(f"Error: {pt1} and {pt2} are in the same group but are significantly different.")

# Function to print Tukey pairwise comparison results
def print_tukey_pair(tukey_df, name1, name2):
    results = tukey_df[((tukey_df['group1_name'] == name1) & (tukey_df['group2_name'] == name2)) |
                       ((tukey_df['group1_name'] == name2) & (tukey_df['group2_name'] == name1))]
    return results


# Create the dictionary of significant differences
significant_diff = create_significant_diff_dict(tukey_results)

# Group property types
grouped_property_types = group_property_types(significant_diff)

# Print the groups
for i, group in enumerate(grouped_property_types):
    print(f"Group {i+1}: {group}")

# Validate the groups
validate_groups(grouped_property_types, tukey_results)

print("\nTukey test results for 'Entire home' vs 'Entire vacation home':")
print(print_tukey_pair(tukey_results, 'Entire home', 'Entire vacation home'))
```

# Final Groups
**Ultimately the property types were placed into 9 different group categories.  The anova and tukey test showed that most of these groups had mean prices that were significantly different from each other.  Only group 8 and group 9 have means that are similar**

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.multicomp import pairwise_tukeyhsd


# Create a new column in the DataFrame for the group
def assign_group(property_type):
    for group, types in My_property_groups.items():
        if property_type in types:
            return group
    return None

df['property_group'] = df['property_type'].apply(assign_group)

# Log transform the price
df['log_price'] = np.log1p(df['price'])  # Using log1p to handle zero prices if any

# Visualize the distribution before and after log transformation
plt.figure(figsize=(12, 5))
plt.subplot(121)
sns.histplot(df['price'], kde=True)
plt.title('Price Distribution (Before Log Transform)')
plt.subplot(122)
sns.histplot(df['log_price'], kde=True)
plt.title('Price Distribution (After Log Transform)')
plt.tight_layout()
plt.show()

# Perform ANOVA on log-transformed prices
anova_model = ols('log_price ~ C(property_group)', data=df).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)
print("ANOVA results (on log-transformed prices):")
print(anova_table)

# Perform Kruskal-Wallis H-test (no need for log transformation here)
grouped = df.groupby('property_group')['price'].apply(list)
kruskal_result = stats.kruskal(*grouped)
print("\nKruskal-Wallis H-test result:")
print(f"H-statistic: {kruskal_result.statistic}")
print(f"p-value: {kruskal_result.pvalue}")

# Visualize log-transformed prices by group
plt.figure(figsize=(12, 6))
sns.boxplot(x='property_group', y='log_price', data=df)
plt.title('Log-transformed Prices by Property Group')
plt.xlabel('Log Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('property_groups_boxplot.png')
plt.show()

# Post-hoc test: Tukey's HSD
tukey_results = pairwise_tukeyhsd(df['price'], df['property_group'])
print("\nTukey's HSD Test Results:")
print(tukey_results)

# Visualize Tukey's HSD results
plt.figure(figsize=(12, 8))
tukey_results.plot_simultaneous()
plt.title("Tukey's HSD Test Results")
plt.xlabel('Price')
plt.savefig('tukey_simultaneous.png')
plt.tight_layout()
plt.show()
```
